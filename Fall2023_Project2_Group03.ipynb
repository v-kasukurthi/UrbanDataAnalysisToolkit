```python
import pandas as pd
CrimeData1=pd.read_csv("/Users/vamshikonapuram/Library/CloudStorage/OneDrive-UMBC/601/Assignments/Project 1/Part_1_Crime_Data.csv")
CrimeData1
```

```python
CrimeData1['CrimeDateTime'] = pd.to_datetime(CrimeData1['CrimeDateTime'], format='%Y/%m/%d %H:%M:%S+%f')
start_date = pd.to_datetime('2017-01-01')
end_date = pd.to_datetime('2022-12-31') 

filtered_data = CrimeData1[(CrimeData1['CrimeDateTime'] >= start_date) & (CrimeData1['CrimeDateTime'] < end_date)]
filtered_data
```

```python
import pandas as pd

filtered_data['Year'] = filtered_data['CrimeDateTime'].dt.year

crime_rate_by_year = filtered_data.groupby('Year')['CrimeDateTime'].count()

overall_change_percentage = crime_rate_by_year.pct_change() * 100
summary_table1 = pd.DataFrame({
    'Year': crime_rate_by_year.index,  # Add the 'Year' column
    'Crime Count': crime_rate_by_year,
    'Overall Change Percentage': overall_change_percentage
})

# Print or display the updated summary table
summary_table1
```

```python
grouped_data = filtered_data.groupby(['Neighborhood', 'Description','Year'])['Total_Incidents'].sum().reset_index()
grouped_data = pd.DataFrame(grouped_data)
grouped_data
```

```python
data_2017 = grouped_data[grouped_data['Year'] == 2017]
data_2022 = grouped_data[grouped_data['Year'] == 2022]
merged_data = pd.merge(data_2017, data_2022, on=['Neighborhood', 'Description'], suffixes=('_2017', '_2022'), how='outer')
merged_data['Change'] = merged_data['Total_Incidents_2022'] - merged_data['Total_Incidents_2017']
merged_data[['Neighborhood', 'Description', 'Total_Incidents_2017', 'Total_Incidents_2022', 'Change']]
```

```python
changes_by_neighborhood = merged_data.groupby('Neighborhood')['Change'].sum().reset_index()
changes_by_neighborhood
```
